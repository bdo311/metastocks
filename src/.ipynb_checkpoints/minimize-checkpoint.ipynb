```python
import csv
import numpy as np
import pandas as pd
import sklearn
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from scipy.optimize import minimize
%matplotlib inline
```

## Get metagene output

```python
category_to_vals = {}
for i in range(1,10):
    fn = glob.glob("../data/metagene_output/no-market-cap/{}*".format(i))[0]
    data = pd.read_csv(fn, header=False, index_col=0)
    category_to_vals[i] = data.loc["weighted"]
category_table = pd.DataFrame.from_dict(category_to_vals)
category_table.head()
```

```python
category_table08 = category_table.loc[category_table.index > '2008-01-01']
category_table08.head()
```

```python
category_table08_20 = category_table08.iloc[range(0,category_table08.shape[0],20),:]
category_table08_20.head()
```

```python
plt.figure(figsize=[15,2.2])
labelpts = range(0, category_table08_20.shape[0], 20)
labels = category_table08_20.index[labelpts]
names = ["basic", "conglom", "consumer", "financial", "health", "industrial", "services", "tech", "utilities"]
for i in range(1,10):
    ax = plt.subplot(2,5, i)
    plt.plot(category_table08_20[names[i-1]])
    plt.title("Category {}: {}".format(i, names[i-1]))
    ax.xaxis.set_ticklabels([])
    #plt.xticks(labelpts, labels, rotation=45)
plt.tight_layout()
plt.savefig("../figures/category_sparklines.pdf")
```

```python
def func(x, alpha, y, A):
    return np.linalg.norm(y - A.dot(x))**2 + 0.5 * alpha * np.linalg.norm(x)**2

def func_deriv(x, alpha, y, A):
    deriv = -2 * y.T.dot(A) + 2 * A.T.dot(A).dot(x) + alpha * np.array(x)
    return list(deriv)

cons = ({'type': 'eq',
         'fun': lambda x: np.sum(x) - 1,
         'jac': lambda x: [1] * len(x)})

def plot_decomp(fn, stockname, row, alpha=1, window=25):
    stock = pd.read_csv(fn, index_col=0)
    stock08 = stock.loc[stock.index > '2008-01-01']["Close"]
    stock08_20 = stock08.iloc[range(0,stock08.shape[0],20)]/stock08.loc['2009-03-06']
    stock08_20 = stock08_20.iloc[:category_table08_20.shape[0]]
    stock08_20.head()
    
    labelpts = range(0, category_table08_20.shape[0], 20)
    labels = category_table08_20.index[labelpts]
    plt.subplot(5, 2, row*2 - 1)
    plt.plot(stock08_20)
    plt.xticks(labelpts, labels, rotation=45)
    plt.ylabel("Normalized price")
    plt.title(stockname)
    
    resdict = {}
    for i in range(category_table08_20.shape[0]-window):
        A = np.matrix(category_table08_20.iloc[i:(i+window),:])
        A -= np.median(A, axis=0)
        y = np.array(stock08_20[i:(i+window)])
        y -= np.median(y)
        
        x = minimize(func, [1/9]*9, args=(alpha,y,A,), jac=func_deriv, method="SLSQP", bounds = [(0,1)]*9, constraints=cons).x
        resdict[i] = [round(j,2) for j in x]
        
    df = pd.DataFrame.from_dict(resdict).transpose()
    df.columns=["basic", "conglom", "consumer", "financial", "health", "industrial", "services", "tech", "utilities"]
    
    
    labelpts = range(0 + window/2, category_table08_20.shape[0] - window/2, 10)
    labels = category_table08_20.index[labelpts]
    max_prop = df.apply(np.max, axis=0)
    max_prop.sort(ascending=False)
    plt.subplot(5, 2, row*2)
    for i in range(5):
       plt.plot(df[max_prop.index[i]], label=max_prop.index[i])   
    plt.legend()
    plt.title("Breakdown of {}".format(stockname))
    plt.xticks(np.array(labelpts) - window/2, labels, rotation=45)
    plt.ylabel("Proportion of price explained")
    plt.tight_layout()
```

```python
plt.figure(figsize=[11,13])
plot_decomp("../data/historical-data/9-utilities/PCG_from_google.csv", "PG&E -- PCG (Utilities)", row=1, alpha=5, window=30)
plot_decomp("../data/historical-data/4-financials/DFS_from_google.csv", "Discover -- DFS (Financial)", row=2, alpha=5, window=30)
plot_decomp("../data/historical-data/4-financials/C_from_google.csv", "Citigroup -- C (Financial)", row=3, alpha=5, window=30)
plot_decomp("../data/historical-data/8-technology/ADBE_from_google.csv", "Adobe Systems -- ADBE (Tech)", row=4, alpha=5, window=30)
plot_decomp("../data/historical-data/8-technology/IBM_from_google.csv", "IBM (Tech)", row=5, alpha=5, window=30)
plt.savefig("../figures/breakdowns.pdf")
```
